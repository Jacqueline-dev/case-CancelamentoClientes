```python
import pandas as pd
tabela = pd.read_csv("/content/cancelamentos.csv")
display(tabela)
```

```python
# Tratamento de erros
display(tabela.info())
```

```python
# Tratar valores vazios
tabela = tabela.dropna()
```

```python
# Vericando se as linhas vazias foram deletadas
display(tabela.info())
```

```python
# Excluir coluna que não faz diferente pra analise de dados
tabela = tabela.drop("CustomerID", axis=1)
display(tabela)
```

```python
# Análise Inicial de Dados
display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
display(tabela.groupby("duracao_contrato").count())
```

```python
display(tabela.groupby("duracao_contrato").mean())
display(tabela.groupby("assinatura").mean())
```

```python
# Contrato mensal sempre cancela
tabela = tabela[tabela["duracao_contrato"]!="Monthly"]

# Vamos ver o cancelamento tirando a assinatura mensal
display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
# Anáise Profunda da base de dados (encontrando a causa dos cancelamentos)
import plotly.express as px

for coluna in tabela.columns:
  grafico = px.histogram(tabela, x="assinatura", color="cancelou")
  grafico.show()
```

```python
# Ligações call center a partir de 5, cancelam
tabela = tabela[tabela["ligacoes_callcenter"]< 5]

# Dias de atraso a partir de 21 cancelam
tabela = tabela[tabela["dias_atraso"] <= 20]


# Vamos ver o cancelamento tirando a assinatura mensal
display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
# Motivos do cancelamento:
# Formas de pagamentos - mensal;
# Dias de atraso > 20;
# Ligações do call center >= 5
```
